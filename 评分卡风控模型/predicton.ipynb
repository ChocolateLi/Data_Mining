```python
# 筛选特征值。并不是每个特征我们都需要，我们需要IV>0.1的特征字段
'''
df:数据集
feature:计算IV值是一个一个特征进行计算
'''
import numpy as np

def call_IV(df,feature,target):
    cols = ['Variable','Value','All','Bad']
    lst = []
    # 找到feature中不同的取值
    for i in range(df[feature].nunique()):
        val = list(df[feature].unique())[i]
        # print(val)
        # 统计feature，feature_value,value的个数，这个value导致target=1的个数
        temp1 = df[df[feature]==val].count()[feature] # 总数
        temp2 = df[(df[feature]==val) & (df[target]==1)].count()[feature] # target=1的个数
        # print(val,temp1,temp2)
        lst.append([feature,val,temp1,temp2])
    # 计算字段 计算WOE和IV值的特征
    data = pd.DataFrame(lst,columns=cols)
    # return data
    data = data[data['Bad']>0]
    data['Margin Bad'] = data['Bad']/data['Bad'].sum()
    data['Margin Good'] = (data['All']-data['Bad']) / (data['All'].sum()-data['Bad'].sum())
    data['WOE'] = np.log(data['Margin Bad']/data['Margin Good'])
    data['IV'] = (data['Margin Bad']-data['Margin Good'])*data['WOE']
    return data['IV'].sum()

call_IV(df_train,'bin_age','SeriousDlqin2yrs')
```

```python
# 计算字段的WOE特征值
def call_WOE(df,features,target):
    for feature in features:
        # sum代表违约个数，count代表一共有多少人
        df_woe = df.groupby(feature).agg({target:['sum','count']})
        # map() 会根据提供的函数对指定序列做映射。第一个参数 function 以参数序列中的每一个元素调用 function 函数，返回包含每次 function 函数返回值的新列表
        df_woe.columns = list(map(''.join,df_woe.columns.values))
        df_woe = df_woe.reset_index()
        df_woe.rename(columns={target+'sum':'bad',target+'count':'all'},inplace=True)
        df_woe['good'] = df_woe['all']-df_woe['bad']
        df_woe['Margin Bad'] = df_woe['bad']/df_woe['bad'].sum()
        df_woe['Margin Good'] = df_woe['good']/df_woe['good'].sum()
        # 为了避免分母为0，使用np.log1p
        df_woe['woe'] = np.log1p(df_woe['Margin Bad']/df_woe['Margin Good'])
        df_woe.columns = [c if c==feature else c+'_'+feature for c in df_woe.columns]
        # 拼接
        df = df.merge(df_woe,on=feature,how='left')
    return df

df_woe = call_WOE(df_train,['bin_age'],'SeriousDlqin2yrs')
df_woe
```
